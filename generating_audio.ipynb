## function

```python
def generate(encoder_model_path, synthesizer_model_path, vocoder_model_path, audio_file_path, text):
    
    #imports
    from pathlib import Path
    import librosa
    import numpy as np
    import soundfile as sf
    import torch
    
    import librosa.display 
    import IPython.display as ipd
    import matplotlib.pyplot as plt
    
    #inference modules
    from encoder import inference2 as encoder
    from synthesizer.inference import Synthesizer
    from vocoder import inference as vocoder
    from encoder.audio import preprocess_wav

    #loading models
    encoder.load_model(encoder_model_path)
    synthesizer = Synthesizer(synthesizer_model_path)
    synthesizer.load()
    vocoder.load_model(vocoder_model_path)
    
    
    original_wav, sampling_rate = librosa.load(audio_file_path)
    preprocessed_wav = preprocess_wav(original_wav, sampling_rate)
    
    embed = encoder.embed_utterance(preprocessed_wav, using_partials = False)
    print("Created the embedding")  

   
    texts = [text]
    embeds = [embed]
    
    specs = synthesizer.synthesize_spectrograms(texts, embeds)
    spec = specs[0]

    print("Created the mel spectrogram")
   
    print("Synthesizing the waveform:")
    
    generated_wav = vocoder.infer_waveform(spec)

    return generated_wav
```

```python
from pathlib import Path
import IPython.display as ipd
saved_encoder_model_path = Path("C:/Users/Swaroop/OneDrive/Documents/Real-Time-Voice-Cloning-master/saved_models/default/encoder.pt")
saved_sythesizer_model_path = Path("./synthesizer/synthesizer.pt")
saved_vocoder_model_path = Path("./vocoder/vocoder.pt")

text = "hello my name is ananth. i am a NLP trainee"
audio_file_path = Path("C:/Users/Swaroop/Downloads/both.flac")

generated_wav = generate(saved_encoder_model_path, saved_sythesizer_model_path, saved_vocoder_model_path, audio_file_path, text )
ipd.Audio(generated_wav, rate = 16000)
        
```
